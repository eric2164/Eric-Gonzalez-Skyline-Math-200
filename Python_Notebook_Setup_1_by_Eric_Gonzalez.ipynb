Research Question #1 : Is there an association between the amount of times eating fast food in the past 7 days and hours spent outside during a typical summer week?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1: 

image.png

```python
#Manage data for variable 1
```

```python
df['H4GH8'].dtype
```

```python
df['Fast_food_consumed'] = df['H4GH8']
df['Fast_food_consumed'].dtype
```

```python
df['Fast_food_consumed'] = pd.to_numeric(df['Fast_food_consumed'], errors = 'coerce')
df['Fast_food_consumed'].dtype
```

```python
df['Fast_food_consumed'] = df['Fast_food_consumed'].loc[ df['Fast_food_consumed'] < 996]
df['Fast_food_consumed'] = df['Fast_food_consumed'].loc[df['Fast_food_consumed'] != ' ']
df['Fast_food_consumed'] = df['Fast_food_consumed'].loc[df['Fast_food_consumed'] != '.']
df['Fast_food_consumed'].max()
```
